```python
import math
from tqdm import tqdm
from sympy import totient
import matplotlib.pyplot as plt
```

$\textbf{Project Euler 73: Counting Fractions in a Range Solution}$

A brute force solution would be to check every numerator $\frac{d}3<n<\frac{d}2$ for each $d\leq12000$ and count those which are coprime.

There will be a much faster way, but I have not found it...

```python
def brute_force_answer(d_max):
    total = 0
    for d in range(4, d_max + 1):
        for n in range(math.ceil(d/3), d//2 + 1):
            if math.gcd(d, n) == 1:
                total += 1
    return total


print("The answer is {}".format(brute_force_answer(12000)))
```

Attempt at a better solution:

Let $F(d,m):=\#\{n: (n,d)=1, 1\leq n\leq m\}$ be the 'windowed $\phi$' function. This function is roughly:
\begin{align*}
    F(d,m)&=\sum_{n=1}^m\mathbb{1}_{(d,n)=1}\\
    &=\sum_{n=1}^m\sum_{k|(n,d)}\mu(k)\\
    &=\sum_{k|d}\left\lceil\frac{m}k\right\rceil\mu(k)\\
    &\approx m\sum_{k|d}\frac{\mu(k)}k\\
    &=\frac{\phi(d)}dm
\end{align*}
So we should expect the number of reduced fractions in this interval to be
\begin{align*}
    T&=\sum_{d=4}^{12000}\left[ F\left(d, \left\lfloor\frac{d}2\right\rfloor\right)-F\left(d, \left\lceil\frac{d}3\right\rceil-1\right)\right]\\
    &=\sum_{d=4}^{12000}\frac{\phi(d)}d\left(\left\lfloor\frac{d}2\right\rfloor-\left\lceil\frac{d}3\right\rceil+1\right)
\end{align*}

```python
def approximation(m):
    total = 0
    for d in range(4, m+1):
        total += ((d//2 - math.ceil(d/3)+1)/d)*totient(d)
    return total


r = range(4, 200)
l = []
for m in r:
    l.append(approximation(m)-brute_force_answer(m))
    
plt.scatter(r, l)
plt.show()
```

The error term here seems to be roughly linear. I can't find a closed-form expression so the brute-force solution will have to do for now!
